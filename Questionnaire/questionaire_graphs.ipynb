# Questionnaire highlights

This notebook aims at showing the highlights of the questionnaire.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("Counter-Strike-questionnaire.csv")
```

```python
premier_elo = df['What is your premier elo?']
cheater_freq = df['How often do you experience playing in a match with a cheater?']
years_played = df['For how many years have you played Counter-Strike']
hours_played = df['How many hours have you played the game']

mapping = {"Never (0% of the time)": 0,
           "Occasionally": 1,
           "Sometimes  (50% of the time)": 2,
           "Often": 3,
           "Always (100% of the time)": 4}
cheater_freq = [mapping[x] for x in cheater_freq]
```

# Cheaters in high elo

One of the notions when reading through the questionnaire answers is that cheaters are more frequent in higher elo games. To test whether this is true we can plot how often users encounter cheaters vs their elo.

```python
plt.title("Cheater Frequency vs Premier ELO")
plt.xlabel("Premier ELO")
plt.ylabel("Cheater Frequency")
plt.yticks(np.arange(0, 5, 1), ["Never (0% of the time)", "Occasionally", "Sometimes (50% of the time)", "Often", "Always (100% of the time)"])
plt.scatter(premier_elo, cheater_freq, c='blue', alpha=0.5)
plt.show()
```

```python
plt.title("Cheater Frequency vs hours played")
plt.xlabel("Hours Played")
plt.ylabel("Cheater Frequency")
plt.yticks(np.arange(0, 5, 1), ["Never (0% of the time)", "Occasionally", "Sometimes (50% of the time)", "Often", "Always (100% of the time)"])
plt.scatter(hours_played, cheater_freq, c='blue', alpha=0.5)
plt.show()
```

```python
plt.title("Cheater Frequency vs Years Played")
plt.xlabel("Years Played")
plt.ylabel("Cheater Frequency")
plt.yticks(np.arange(0, 5, 1), ["Never (0% of the time)", "Occasionally", "Sometimes (50% of the time)", "Often", "Always (100% of the time)"])
plt.scatter(years_played, cheater_freq, c='blue', alpha=0.5)
plt.show()
```
